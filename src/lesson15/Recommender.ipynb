```python
import json 
with open('movies.small.json') as f: 
    reviews = json.load(f)
```

```python
reviews = map(lambda x: {'productId': x['productId'], 'score': x['score'], 'profileName': x['profileName']}, reviews)
```

```python
reviews[-10:]
```

```python
productIds = list({x['productId'] for x in reviews})
profileNames = list({x['profileName'] for x in reviews})
```

```python
matdicts = []
for i, profileName in enumerate(profileNames): 
    d = {}
    for rev in reviews: 
        if rev['profileName'] == profileName:
            d[rev['productId']] = rev['score']
    matdicts.append(d)
```

```python
matdicts[:10]
```

```python
from sklearn.feature_extraction import DictVectorizer 

vect = DictVectorizer()

R = vect.fit_transform(matdicts)
```

```python
from collections import defaultdict
empirical_means = defaultdict(float)
for i in xrange(R.shape[0]):
    for j, v in zip(R[i].indices, R[i].data): 
        empirical_means[i] += v
    empirical_means[i] = empirical_means[i]/R[i].nnz
print(empirical_means)
```

```python
R = R.todense()

for i in xrange(R.shape[0]): 
    for j in xrange(R.shape[1]): 
        if R[i,j] == 0: 
            R[i,j] = empirical_means[i]
print R
```

```python
Rtilde = R - R.mean(axis=1)
```

```python
from sklearn.decomposition import TruncatedSVD 

N_COMPONENTS = 100

tsvd = TruncatedSVD(N_COMPONENTS)

R_reduced = tsvd.fit_transform(Rtilde)
```

```python
R_reduced
```

```python
import sklearn.metrics.pairwise as smp 
similarities = smp.cosine_similarity(R_reduced)
```

```python
idx = 1028
TOP_N = 10

top_n = {}

top_n[idx] = similarities[idx].argsort()[-TOP_N-1:-1]
```

```python
top_n[idx]
```

```python
col_idx = 15

rating = 0.
for sim_idx in top_n[idx]: 
    rating += Rtilde[sim_idx, col_idx]
rating = rating/TOP_N
```

```python
import numpy as np
means = np.array(R.mean(axis=1)).flatten()
```

```python
rating += means[idx]
```

```python
Rtilde_reconstructed = np.dot(R_reduced, tsvd.components_)
```
